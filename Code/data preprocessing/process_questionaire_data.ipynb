```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "small_data.csv"
df = pd.read_csv(file_path, encoding='utf-8-sig')
df.head()
```

```python
# 获取所有列名
all_columns = df.columns.tolist()

# 前4列和后11列的列名
cols_to_drop = all_columns[:4] + all_columns[-11:]

# 删除这些列
df = df.drop(columns=cols_to_drop)
df.head()
```

```python
df.rename(columns={'1.性别': 'gender'}, inplace=True)
df['gender'] = df['gender'].map({'A.女': 0, 'B.男': 1})
df.head()
```

```python
df.rename(columns={'2.年龄': 'age'}, inplace=True)
df['age'] = df['age'].str[0]
df.head()
```

```python
df.rename(columns={'3.饮食偏好:是否喜欢辣': 'like_spicy'}, inplace=True)
df.rename(columns={'3.饮食偏好:是否喜欢甜': 'like_sweet'}, inplace=True)
df.rename(columns={'3.饮食偏好:是否喜欢酸': 'like_sour'}, inplace=True)
df.rename(columns={'3.饮食偏好:是否偏爱素食': 'like_veg'}, inplace=True)
df.rename(columns={'3.饮食偏好:能否接受生食（如刺身）': 'accept_raw'}, inplace=True)
df.rename(columns={'3.饮食偏好:能否接受冷食（如冰淇淋）': 'accept_cold'}, inplace=True)
df.rename(columns={'3.饮食偏好:能否接受带有臭味的食物（如螺蛳粉 榴莲 臭豆腐）': 'accept_stint'}, inplace=True)
df.rename(columns={'3.饮食偏好:能否接受重口味食物': 'accept_hard'}, inplace=True)
df.rename(columns={'3.饮食偏好:能否接受清淡口味食物': 'accept_light'}, inplace=True)
df['like_spicy'] = df['like_spicy'].map({'A.是': 1, 'B.否': 0})
df['like_sweet'] = df['like_sweet'].map({'A.是': 1, 'B.否': 0})
df['like_sour'] = df['like_sour'].map({'A.是': 1, 'B.否': 0})
df['like_veg'] = df['like_veg'].map({'A.是': 1, 'B.否': 0})
df['accept_raw'] = df['accept_raw'].map({'A.是': 1, 'B.否': 0})
df['accept_cold'] = df['accept_cold'].map({'A.是': 1, 'B.否': 0})
df['accept_stint'] = df['accept_stint'].map({'A.是': 1, 'B.否': 0})
df['accept_hard'] = df['accept_hard'].map({'A.是': 1, 'B.否': 0})
df['accept_light'] = df['accept_light'].map({'A.是': 1, 'B.否': 0})
df.head()
```

```python
df.rename(columns={'4.您喜欢的菜系（多选）:川菜（如麻辣香锅、毛血旺）': 'Chuan'}, inplace=True)
df.rename(columns={'4.您喜欢的菜系（多选）:粤菜&港式（如烧腊、菠萝油）': 'Yue'}, inplace=True)
df.rename(columns={'4.您喜欢的菜系（多选）:湘菜（如小炒黄牛肉）': 'Xiang'}, inplace=True)
df.rename(columns={'4.您喜欢的菜系（多选）:江浙菜（如西湖醋鱼、东坡肉）': 'Jiangzhe'}, inplace=True)
df.rename(columns={'4.您喜欢的菜系（多选）:日料（如寿司、刺身）': 'Japanese'}, inplace=True)
df.rename(columns={'4.您喜欢的菜系（多选）:韩餐（如石锅拌饭）': 'Korean'}, inplace=True)
df.rename(columns={'4.您喜欢的菜系（多选）:西餐（如牛排、披萨）': 'Western'}, inplace=True)
df.rename(columns={'4.您喜欢的菜系（多选）:东南亚风格（如冬阴功汤）': 'Southeast_asia'}, inplace=True)
columns_todrop = [
    '4.您喜欢的菜系（多选）:其他____{fillblank-514a}',
    '4.您喜欢的菜系（多选）:其他____{fillblank-514a}[选项填空]'
]
df.drop(columns=columns_todrop, inplace=True, errors='ignore') 
df['Chuan'] = np.where(df['Chuan'].notna(), 1, 0)
df['Yue'] = np.where(df['Yue'].notna(), 1, 0)
df['Xiang'] = np.where(df['Xiang'].notna(), 1, 0)
df['Jiangzhe'] = np.where(df['Jiangzhe'].notna(), 1, 0)
df['Japanese'] = np.where(df['Japanese'].notna(), 1, 0)
df['Korean'] = np.where(df['Korean'].notna(), 1, 0)
df['Western'] = np.where(df['Western'].notna(), 1, 0)
df['Southeast_asia'] = np.where(df['Southeast_asia'].notna(), 1, 0)
df.head()
```
